```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

sns.set_style("darkgrid")
df=pd.read_csv("pca.csv")
print(df.head())
df.plot(x="nr_comp", y="cummulated_std_dev")
plt.show()
```
