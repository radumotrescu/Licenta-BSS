```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

sns.set_style("darkgrid")
df=pd.read_csv("pca.csv")
print(df.head())
df.plot(x="nr_comp", y="cummulated_std_dev")
plt.show()
```

```python
df = pd.DataFrame({"nr_comp": [1,2,3,4,5],
                   "std_dev": [80,10,5,3,2]})

ax = df.plot(x="nr_comp", y="std_dev", legend=False)
ax.figure.legend()
plt.show()
```
